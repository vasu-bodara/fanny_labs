```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("D:/python/ev.csv")
```

```python
print("Dataset Preview:")
print(df.head())
print("\nDataset Info:")
print(df.info())
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=df['Manufacturer'], hue=df['Vehicle Type'], palette='coolwarm')
plt.title("Model - Manufacturer - Vehicle Type Distribution")
plt.xlabel("Count")
plt.ylabel("Manufacturer")
plt.legend(title="Vehicle Type")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(y=df['Model'], x=df['Price'], palette='Blues_r')
plt.xlabel("Price (INR)")
plt.ylabel("EV Model")
plt.title("Model vs. Price")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(y=df['Model'], x=df['Range per Charge (km)'], palette='magma')
plt.xlabel("Range per Charge (km)")
plt.ylabel("EV Model")
plt.title("Range vs. EV Models")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Vehicle Type'], y=df['Battery Capacity (kWh)'], palette='Set2')
plt.xlabel("Vehicle Type")
plt.ylabel("Battery Capacity (kWh)")
plt.title("Vehicle Type vs. Battery Capacity")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['Charging Time'], y=df['Range per Charge (km)'], hue=df['Model'], palette='viridis', s=100)
plt.xlabel("Charging Time (hours)")
plt.ylabel("Range per Charge (km)")
plt.title("Model - Charging Time vs. Range")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=df['Manufacturer'], hue=df['Year of Manufacture'], palette='muted')
plt.xlabel("Count")
plt.ylabel("Manufacturer")
plt.title("Manufacturer - Year of Manufacture Distribution")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['Battery Capacity (kWh)'], y=df['Charging Time'], hue=df['Model'], palette='coolwarm', s=100)
plt.xlabel("Battery Capacity (kWh)")
plt.ylabel("Charging Time (hours)")
plt.title("Battery Capacity vs. Charging Time")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
vehicle_type_demand = df['Vehicle Type'].value_counts()
vehicle_type_demand.plot(kind='bar', colormap='viridis')
plt.title("Vehicle Type Demand")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Models")
plt.xticks(rotation=45)
plt.show()
```

```python
features = ["Battery Capacity (kWh)", "Range per Charge (km)", "Charging Time", "Price", "Power (HP or kW)", "Top Speed (km/h)"]
X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster_KMeans'] = kmeans.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['Power (HP or kW)'], y=df['Range per Charge (km)'], hue=df['Cluster_KMeans'], palette='coolwarm', s=100)
plt.xlabel("Power (HP or kW)")
plt.ylabel("Range per Charge (km)")
plt.title("Market Segmentation using K-Means Clustering")
plt.show()
```
